Comando básicos

*   "r" -> read -> ler (recomendado para poucas informação)
* readline -> para arquivos maiores
*   "w" -> write -> escrever
* "a" -> append -> adicionar



```python
#Como o arquivo se encontra no mesmo local do ambiente do codigo
#precisamos apenas escrever o nome dele, caso opte por colocar em uma pasta especificar o caminho


#utilizando try e except para verificar se o arquivo foi lido e salvo corretamente
#primeiro parametro passado é o nome ou caminho do arquivo
#segundo é o metodo "r" de read (ler)
#e armazenando o conteudo que se encontra em uma variavel
try:
    with open("consultores.txt", "r", encoding="utf-8") as arquivo:
        dataSet = arquivo.readlines()
    print("Leitura bem-sucedida")
except IOError as e:
    print(f"Ocorreu um erro durante a leitura do arquivo: {e}")
```

# Salvando os dados em um novo arquivo

```python
#utilizando a biblioteca datetime
#com a funcao datetime.now(que obtém a data e a hora atuais)
#e o strftime é um método que formata o objeto  retornado pelo datetime.now() em uma string
#No formato passado "%m" que representa o mês do ano
data_atual = datetime.now().strftime("/%m")
data_atual_arq = datetime.now().strftime("%m")

# Formata o nome do arquivo de aniversariantes do mês com a data atual
nome_arquivo_aniversariantes = "Aniversariantes_do_mês_" + data_atual_arq +".txt"

#o argumento "w" no permite substituir os conteudo do arquivo pelo que queremos
#caso o nome do arquivo passado não exista, ele ira criar um novo arquivo com esse nome salvar nele as informações que queremos
#em seguida utilizamos um for para percorrer linha por linha da lista dataSet
#onde contem os dados dos consultores e verificar na linha se encontra a data_atual
#se existir iremos guarda os dados da linha inteira em um novo arquivo.txt

try:
    with open(nome_arquivo_aniversariantes, "w", encoding="utf-8") as aniversariantes:
        for linha in dataSet:
            if data_atual in linha:
                aniversariantes.write(linha)
    print("Escrita bem-sucedida")
except IOError as e:
    print(f"Ocorreu um erro durante a escrita do arquivo: {e}")
```

# Testando

```python
# testando
#lendo o arquivo que foi salvo com o dados dos aniversariantes e imprimindo
try:
    with open(nome_arquivo_aniversariantes, "r", encoding="utf-8") as aniversariantes:
        aniversariantes_lst = aniversariantes.readlines()
        for linha in aniversariantes_lst:
            print(linha)
    print("Leitura bem-sucedida")
except IOError as e:
    print(f"Ocorreu um erro durante a leitura do arquivo: {e}")
```
